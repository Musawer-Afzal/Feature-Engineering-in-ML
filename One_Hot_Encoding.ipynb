## One Hot Encoding Using SKLearn One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialise the One Hot Encoder
ohe = OneHotEncoder()

# Identify the categorical columns to be encoded
col_names = ['neighbourhood_group']

# Apply the OHE encoding
dummy_cols = ohe.fit_transform(df[col_names])

# Get output feature names correctly
dummy_names = ohe.get_feature_names_out(col_names)

# Create a DataFrame of the encoded columns
dummy_cols = pd.DataFrame(dummy_cols.toarray(), columns=dummy_names, dtype=int, index=df.index)

# Combine with the rest of the data
df_ohe_encoded = pd.concat([df.drop(columns=col_names), dummy_cols], axis=1)

df_ohe_encoded
```

```python
col_names = ['node_caps']

# Fit the encoder
ohe = OneHotEncoder(sparse_output=False)  # use sparse_output=False if you want a dense array directly
bc_prefix = ohe.fit_transform(bc_df[col_names])

# Rename the columns with prefix 'nc'
categories = ohe.categories_[0]
bc_prefix_names = [f"nc_{cat}" for cat in categories]

# Create the DataFrame
bc_prefix = pd.DataFrame(bc_prefix, columns=bc_prefix_names, dtype=int, index=bc_df.index)

# Combine with original DataFrame
bc_df_ohe = pd.concat([bc_df.drop(columns=col_names), bc_prefix], axis=1)

bc_df_ohe
```
