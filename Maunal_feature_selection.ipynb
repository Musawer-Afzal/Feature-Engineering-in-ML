```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_cars = pd.read_csv("Data-Prep-for-ML/Feature Selection/indian cars dataset incnulls.csv")
df_cars.head()
```

```python
# One of the method is to check if the feature is giving as any predictive power
# Ending price does not give us any power over the starting price
df_cars = df_cars.drop(columns=['ending_price'])
df_cars.head()
```

```python
# Another method is to check if a feature have too many unique values
df_cars.nunique()
```

```python
# car_name have too many values so less predictive power
df_cars = df_cars.drop(columns=['car_name'])
df_cars.head()
```
