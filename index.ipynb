# Identify Errors for Datatypes

```python
# OH_participated is a object instead of number value, lets fix that
df_grades.info()
df_grades.head()
```

```python
# The below given command can only be run one time or it will give error
# df_grades['OH_participated'] = df_grades['OH_participated'].str.replace('"', '').astype(float)
df_grades.info()
```

```python
df_grades.isna().sum()
```

```python
df_grades['OH_participated'] = df_grades['OH_participated'].fillna(0)
df_grades['classes_skipped'] = df_grades['classes_skipped'].fillna(0)
df_grades.isna().sum()
```

```python
df_grades.head()
```

```python
# in order to deal with NaN values for tuition
df_grades['tuition'] = df_grades['tuition'].fillna(df_grades['tuition'].mean())
df_grades.isna().sum()
```
